## Create X and y.
### get 10 generations every time 
### an assumption that our generations
### have some relations so our model
### could potantially learn those relations.

```python
def pick_n_gen_slices(df, n=1, ten_gen_start=0):
    df_ps_1 = df.copy()
    n = n * 500
    if(ten_gen_start!=0):
        ten_gen_start = ten_gen_start * n
        
    return df_ps_1[0+ten_gen_start:n-1+ten_gen_start+1]

def create_data(df_ps, n=10, pop_size=100, isSplitted=False):
    new_df = df_ps.copy() 
    data = []
    k= pop_size//n
    if(isSplitted == True):
        # the split is always 0.8 0.2
        # 500 gen *0.8 = 400 gen
        # n = 10 --> 40 samples
         
        #print(pop_size)
        k = (new_df.shape[0] // pop_size) // 5   
    
    for i in range(k):
        my_df = pick_n_gen_slices(new_df, n=n, ten_gen_start = i)
        data.append(my_df.to_numpy())
    data = np.array(data)
    return data
```
